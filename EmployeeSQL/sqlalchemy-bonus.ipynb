```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create engine and connection
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5432/sql_challenge_db')
connection = engine.connect()
```

```python
# Reading employees table
employees_df = pd.read_sql("select * from employees", connection)
employees_df.head()
```

```python
# Reading salaries table
salaries_df = pd.read_sql("select * from salaries", connection)
salaries_df.head()
```

```python
# Reading titles table
titles_df = pd.read_sql("select * from titles", connection)
titles_df.head()
```

```python
# Reading department tables
depts_df = pd.read_sql("select * from departments", connection)
depts_df.head()
```

```python
# Reading dept employees table
dept_emp_df = pd.read_sql("select * from dept_emp", connection)
dept_emp_df.head()
```

```python
# Reading dept managers table
dept_manager_df = pd.read_sql("select * from dept_manager", connection)
dept_manager_df.head()
```

```python
# Histogram of salaries
salaries_df.hist('salary')
plt.title('Employee Salaries Distribution')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()
```
